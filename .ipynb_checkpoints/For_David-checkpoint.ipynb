```python
'''
Import packages
'''

import  scipy.signal.signaltools
%matplotlib notebook

def _centered(arr, newsize):
    # Return the center newsize portion of the array.
    newsize = np.asarray(newsize)
    currsize = np.array(arr.shape)
    startind = (currsize - newsize) // 2
    endind = startind + newsize
    myslice = [slice(startind[k], endind[k]) for k in range(len(endind))]
    return arr[tuple(myslice)]

scipy.signal.signaltools._centered = _centered

import dtscalibration

import os
import warnings

warnings.simplefilter('ignore')  # Hide warnings to avoid clutter in the notebook

from dtscalibration import suggest_cable_shift_double_ended
from dtscalibration import read_silixa_files
from matplotlib import pyplot as plt
import numpy as np
from dtscalibration import shift_double_ended

%matplotlib inline
```

```python
'''
Correct the DTS calibration
'''

filepath = ''

ds_ = read_silixa_files(
    directory=filepath,
    timezone_netcdf='UTC',
    file_ext='*.xml')

ds = ds_.sel(x=slice(0, 628.2))  # only calibrate parts of the fiber


sections = {
    'probe1Temperature': [slice(130., 158.)],  # temp isolator
    }
ds.sections = sections

suggested_shift = suggest_cable_shift_double_ended(
    ds,
    np.arange(-5, 5),
    plot_result=True,
    figsize=(12,8))

ds_restored = shift_double_ended(ds, suggested_shift[0])

print(ds.x, 3*'\n', ds_restored.x)
```

```python
'''
This is now doing the calibration
'''
matching_sections = [
    (slice(130., 158.), slice(470.5, 498.75), False)
]
sections = {
    'probe1Temperature': [slice(130., 158.), slice(470., 500.)]
    }
ds_.sections = sections
st_var, resid = ds.variance_stokes(st_label='st')
ast_var, _ = ds.variance_stokes(st_label='ast')
rst_var, _ = ds.variance_stokes(st_label='rst')
rast_var, _ = ds.variance_stokes(st_label='rast')

ds.calibration_double_ended(
    st_var=st_var,
    ast_var=ast_var,
    rst_var=rst_var,
    rast_var=rast_var,
    trans_att=[314.25],
    matching_sections=matching_sections,
    store_tmpw='tmpw',
    method='wls',
    solver='sparse')

ds_.calibration_double_ended(
    st_var=st_var,
    ast_var=ast_var,
    rst_var=rst_var,
    rast_var=rast_var,
    store_tmpw='tmpw',
    method='wls',
    solver='sparse')
```

### Temperature perturbation at depth $z_j$ is given by $$\theta_j = T_k A_{jk}$$
### $$A_{jk} = erfc{\frac{z_j}{2\sqrt{\kappa t_{k-1}}}} -  erfc{\frac{z_j}{2\sqrt{\kappa t_{k}}}}$$
### where $T_k$ is the surface temperature at time interval k

```python
from scipy.special import erfc
column = borehole_total.x - start_of_borehole
depth_interval = float(ds.attrs['stepIncrement:#text'])
thickness = bottom_trend - start_of_borehole
start = column[0]
stop = column[-1]
kappa = 1e-6*3e7 #thermal diffusivity in years
z = np.arange(start,stop,depth_interval)
t = np.arange(0, 150, .1)
```

```python
A = np.full((len(z),len(t-1)), np.nan)

for z_j in np.arange(0,len(z)):
    for k in np.arange(1,len(t)):
        A[z_j,k] = erfc(z[z_j]/(2*np.sqrt(kappa*t[k-1]))) - erfc(z[z_j]/(2*np.sqrt(kappa*t[k])))
    
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(A[:,:])
plt.show()
```

```python
A_copy = np.nan_to_num(A)
A_copy
```

```python
#Singular value decomposition
u, s, v_T = np.linalg.svd(A_copy, full_matrices=True)
#Plotting the singular values
import matplotlib.pyplot as plt

fig,ax = plt.subplots()

ax.plot(s)
ax.set_xlabel('Index')
ax.set_ylabel('Singular Value')
#plt.xticks(ticks=[0,1,2,3,4])
#plt.xlim([0,5])
plt.show()
```

```python
p = 2
#Applying the thresholding
u_p = u[:,:p]
s_p = np.diag(s[:p])
vh_p = v_T.T[:,:p]

```

```python
print('U_p =\n', u_p)
print('S_p =\n',s_p)
print('V_p =\n',vh_p)
```

```python
#Now finding the generalized G matrix

g_gen = u_p @ s_p @ vh_p.T
print('G = \n', g_gen)
```

```python
u_p_transpose = u_p.T
s_p_inv = np.linalg.inv(s_p)

g_inv = vh_p @ s_p_inv @ u_p_transpose

print('G^-g = \n', g_inv)
```

```python
m,b = np.polyfit(mean_tot.sel(x=slice(top_trend,bottom_trend)).x, mean_tot.sel(x=slice(top_trend,bottom_trend)), 1)

Q_geo = m *2.5*1e3

y  = m*mean_tot.sel(x=slice(start_of_borehole,bottom_trend)).x + b
anomaly = mean_tot.sel(x=slice(start_of_borehole,bottom_trend)) - y
```

```python
anomaly.values
```

```python
m_est = g_inv @ anomaly.values

print('m_est = \n',m_est)
```

```python
import matplotlib.pyplot as plt

plt.plot(m_est)
plt.xlabel('time (before present)')
plt.ylabel('surfae temp anomaly (C)')
#plt.xlim([25,200])
plt.show()
```

```python
uncertainty = np.zeros((z.shape[0]))
uncertainty[:] = 0.01
cov_d = np.diag(uncertainty**2) #assuming that the data is uncorrelated.

print('data covariance matrix, cov_d =  \n', cov_d)

cov_m = g_inv @ cov_d @ g_inv.T

print('Model covariance matrix, cov_m = \n', cov_m)
```

```python
R_model = g_inv @ g_gen
print('Model resolution matrix, R_model = \n',R_model)

model_resolved = np.diag(R_model)
print('Resolution of model parameters: \n', model_resolved)
```

```python
#non-negative least squares routine

from scipy.optimize import nnls

x, rnorm = nnls(A_copy, anomaly.values)

#Finding the L2 norm.
e = anomaly.values - (A_copy @ x) 

l2 = e.T @ e

print("The L2 norm is: ", l2)
```

# Can we do this same SVD in a Monte Carlo style?

```python
from scipy.special import erfc
import matplotlib.pyplot as plt

column = borehole_total.x - start_of_borehole
depth_interval = float(ds.attrs['stepIncrement:#text'])
thickness = bottom_trend - top_trend
distance_of_season = top_trend - start_of_borehole
start = column[int(distance_of_season/depth_interval)+1]
stop = column[-1]
kappa = 1e-6*3e7 #thermal diffusivity in years
z = np.arange(start,stop,depth_interval)
t = np.arange(0, 150, 1)
```

```python
A = np.full((len(z),len(t-1)), np.nan)

for z_j in np.arange(0,len(z)):
    for k in np.arange(1,len(t)):
        A[z_j,k] = erfc(z[z_j]/(2*np.sqrt(kappa*t[k-1]))) - erfc(z[z_j]/(2*np.sqrt(kappa*t[k])))
        

fig, ax = plt.subplots()
ax.plot(A)
plt.show()

A_copy = np.nan_to_num(A)
A_copy
```

```python
#Singular value decomposition
u, s, v_T = np.linalg.svd(A_copy, full_matrices=True)


#Plotting the singular values
fig,ax = plt.subplots(figsize=[5,5])
ax.plot(s)
ax.set_xlabel('Index')
ax.set_ylabel('Singular Value')
#ax.set_xlim([0,10])

#fig.savefig('singular_values.eps')
plt.show()
```

```python
p = 4
#Applying the thresholding
u_p = u[:,:p]
s_p = np.diag(s[:p])
vh_p = v_T.T[:,:p]

print('U_p =\n', u_p)
print('S_p =\n',s_p)
print('V_p =\n',vh_p)
```

```python
#Now finding the generalized G matrix

g_gen = u_p @ s_p @ vh_p.T
print('G = \n', g_gen)

#And the inverse Gen G matrix

u_p_transpose = u_p.T
s_p_inv = np.linalg.inv(s_p)

g_inv = vh_p @ s_p_inv @ u_p_transpose

print('G^-g = \n', g_inv)
```

```python
#remove geo_flux in our data to get the temperature anomaly

m,b = np.polyfit(mean_tot.sel(x=slice(top_trend,bottom_trend)).x, mean_tot.sel(x=slice(top_trend,bottom_trend)), 1)

Q_geo = m *2.5*1e3

y  = m*mean_tot.sel(x=slice(top_trend,bottom_trend)).x + b
anomaly = mean_tot.sel(x=slice(top_trend,bottom_trend)) - y

fig,ax = plt.subplots()
ax.plot(anomaly.values)
ax.set_ylabel('Temperature Anomaly (K)')
ax.set_xlabel('Position')
```

```python
#Monte Carlo
percentage = .30
num = int(len(anomaly.values) * percentage)
iterations = 1000
model_estimates = []
predicted_data = []

for i in range(iterations):
    indicesToRemove = np.random.choice(
                            len(anomaly.values),
                            num,
                            replace=False)
    
    updated_data = np.delete(anomaly.values, indicesToRemove)
    g_inv_updated = np.delete(g_inv, indicesToRemove, 1)
    g_gen_updated = np.delete(g_gen, indicesToRemove, 0)
    
    #model estimates
    m_est = g_inv_updated @ updated_data
    model_estimates.append(m_est)
    
    #predicted data vs actual data

    d_pred = g_gen @  m_est
    predicted_data.append(d_pred)

model_estimates_arr = np.asarray(model_estimates)
predicted_data_arr = np.asarray(predicted_data)

```

```python
monte_model_estimates = model_estimates_arr.T
monte_percentiles = np.percentile(model_estimates_arr.T, [2.5, 50, 97.5], axis=1)

monte_data_preds = predicted_data_arr.T
predicted_data_precentiles = np.percentile(monte_data_preds, [2.5, 50, 97.5], axis=1)
```

```python
fig,(ax1, ax2) = plt.subplots(1,2, figsize=[9,4])
ax1.plot(t, monte_percentiles[1], label='Temperature Inversion')
ax1.fill_between(t, monte_percentiles[0], monte_percentiles[2], color='lightgrey', label='2$\sigma$')
#ax.set_ylabel('Temperature Anomaly (K)')
#ax.set_xlabel('Position')
ax1.grid()
ax1.set_xlim([t[0],t[-1]])
ax1.set_ylabel('Temperature Anomaly (K)')
ax1.set_xlabel('Years (Before Present)')
ax1.annotate('a. Temperature reconstruction.',xy = (0.1,.6), annotation_clip=False, weight='bold')
ax1.legend(loc = 'lower right')


ax2.plot(anomaly.x - top_trend +distance_of_season,anomaly.values, label='Collected Data')
ax2.plot(z,predicted_data_precentiles[1], label='Predicted Data')
ax2.fill_between(z,predicted_data_precentiles[0], predicted_data_precentiles[2], color='lightgrey', label='2$\sigma$')
ax2.grid()
ax2.set_xlabel('Depth (m)')
ax2.annotate('b. Predicted data comparison.',xy = (15,.165), annotation_clip=False, weight='bold')
ax2.legend()

left, bottom, width, height = [0.25, 0.55, 0.2, 0.3]
ax3 = fig.add_axes([left, bottom, width, height])
ax3.plot(t, monte_percentiles[1])
ax3.fill_between(t, monte_percentiles[0], monte_percentiles[2], color='lightgrey')
ax3.set_xlim([0,25])
ax3.set_ylim([-0.25,0.6])


plt.tight_layout()
plt.show()

fig.savefig('temp_reconstruction.eps')
```

# Does lowpass filtering the data improve the results?

```python
from scipy.special import erfc
import matplotlib.pyplot as plt

column = borehole_total.x - start_of_borehole
depth_interval = float(ds.attrs['stepIncrement:#text'])
thickness = bottom_trend - top_trend
distance_of_season = top_trend - start_of_borehole
start = column[int(distance_of_season/depth_interval)+1]
stop = column[-1]
kappa = 1e-6*3e7 #thermal diffusivity in years
z = np.arange(start,stop,depth_interval)
t = np.arange(0, 150, 1)
```

```python
A = np.full((len(z),len(t-1)), np.nan)

for z_j in np.arange(0,len(z)):
    for k in np.arange(1,len(t)):
        A[z_j,k] = erfc(z[z_j]/(2*np.sqrt(kappa*t[k-1]))) - erfc(z[z_j]/(2*np.sqrt(kappa*t[k])))
        

fig, ax = plt.subplots()
ax.plot(A)
plt.show()

A_copy = np.nan_to_num(A)
```

```python
#Singular value decomposition
u, s, v_T = np.linalg.svd(A_copy, full_matrices=True)


#Plotting the singular values
fig,ax = plt.subplots()
ax.plot(s)
ax.set_xlabel('Index')
ax.set_ylabel('Singular Value')
#plt.xticks(ticks=[0,1,2,3,4])
#plt.xlim([0,5])
plt.show()
```

```python
p = 4
#Applying the thresholding
u_p = u[:,:p]
s_p = np.diag(s[:p])
vh_p = v_T.T[:,:p]

print('U_p =\n', u_p)
print('S_p =\n',s_p)
print('V_p =\n',vh_p)
```

```python
#Now finding the generalized G matrix

g_gen = u_p @ s_p @ vh_p.T
print('G = \n', g_gen)

#And the inverse Gen G matrix

u_p_transpose = u_p.T
s_p_inv = np.linalg.inv(s_p)

g_inv = vh_p @ s_p_inv @ u_p_transpose

print('G^-g = \n', g_inv)
```

```python
#remove geo_flux in our data to get the temperature anomaly

m,b = np.polyfit(mean_tot.sel(x=slice(top_trend,bottom_trend)).x, mean_tot.sel(x=slice(top_trend,bottom_trend)), 1)

Q_geo = m *2.5*1e3

y  = m*mean_tot.sel(x=slice(top_trend,bottom_trend)).x + b
anomaly = mean_tot.sel(x=slice(top_trend,bottom_trend)) - y


#lowpass filter
import scipy.signal as signal
sos = signal.butter(4, 20, 'lp', fs=1000, output='sos')
filtered = signal.sosfilt(sos, anomaly.values[:])


fig,ax = plt.subplots()
ax.plot(filtered)
#ax.plot(anomaly.values)
ax.set_ylabel('Temperature Anomaly (K)')
ax.set_xlabel('Position')
```

```python
#Monte Carlo
percentage = .30
num = int(len(filtered) * percentage)
iterations = 10000
model_estimates = []
predicted_data = []

for i in range(iterations):
    indicesToRemove = np.random.choice(
                            len(filtered),
                            num,
                            replace=False)
    
    updated_data = np.delete(filtered, indicesToRemove)
    g_inv_updated = np.delete(g_inv, indicesToRemove, 1)
    g_gen_updated = np.delete(g_gen, indicesToRemove, 0)
    
    #model estimates
    m_est = g_inv_updated @ updated_data
    model_estimates.append(m_est)
    
    #predicted data vs actual data

    d_pred = g_gen @ m_est
    predicted_data.append(d_pred)
    print(i)

model_estimates_arr = np.asarray(model_estimates)
predicted_data_arr = np.asarray(predicted_data)

```

```python
monte_model_estimates = model_estimates_arr.T
monte_percentiles = np.percentile(model_estimates_arr.T, [2.5, 50, 97.5], axis=1)

monte_data_preds = predicted_data_arr.T
predicted_data_precentiles = np.percentile(monte_data_preds, [2.5, 50, 97.5], axis=1)
```
